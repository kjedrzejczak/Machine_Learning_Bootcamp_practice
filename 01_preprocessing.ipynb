```python
import numpy as np
import pandas as pd
import sklearn

print(np.__version__)
print(pd.__version__)
print(sklearn.__version__)
```

##Data

```python
data = {'size' : ['Xl', 'L', 'M', 'L', 'M'],
        'color' : ['red', 'green', 'blue', 'green', 'red'],
        'gender': ['female', 'male', 'male', 'female', 'female'],
        'price': [199.0, 89.0, 99.0, 129.0, 79.0],
        'weight': [500, 450, 300, 300, 410],
        'bought': ['yes', 'no', 'yes', 'no', 'yes']}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

```python
for i in ['size', 'color', 'gender', 'bought']:
  df[i] = df[i].astype('category')

df.info()
```

```python
df.describe().T
```

```python
df.describe(include='category').T
```

```python
df
```

Label_encoding

```python
from  sklearn.preprocessing import LabelEncoder

la = LabelEncoder()
la.fit(df['bought'])
la.transform(df['bought'])
```

```python
la.fit_transform(df['bought'])
```

```python
la.classes_
```

```python
df['bought'] = la.fit_transform(df['bought'])
```

```python
df
```

```python
df['bought'] = la.inverse_transform(df['bought'])
df
```

```python
from sklearn.preprocessing import  OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```
