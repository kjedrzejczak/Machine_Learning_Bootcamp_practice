```python
import numpy as np
import pandas as pd
import sklearn
```

```python
def fetch_financial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')


df_raw = fetch_financial_data()
df_raw.head()
```

```python
df = df_raw.copy()
df = df[:5]
df.info()
```

###Dodawanie zmiennych

```python
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df
```

### Dyskretyzacja zmiennej ciągłej

```python
df['height_cut'] = pd.cut(x=df.High, bins=3)
df
```

```python
df['height_cut'] = pd.cut(x=df.High, bins=(150, 160, 170, 180))
df
```

```python
df['height_cut'] = pd.cut(x=df.High, bins=(150, 160, 170, 180), labels=['small', 'medium', 'high'])
df
```

```python
get_d = df[['High','height_cut']]
get_d
```

```python
pd.get_dummies(get_d,  drop_first=True, prefix='height')
```
